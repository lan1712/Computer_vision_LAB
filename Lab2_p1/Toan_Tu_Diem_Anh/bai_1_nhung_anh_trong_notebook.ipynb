# Bài 1 – Thay đổi độ sáng, tương phản, âm bản, cắt ngưỡng

Mở file là xem được ảnh, không dùng cv2.imshow.

```python

import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python

# Read image
img = cv2.imread('pikachu.jpg')
if img is None:
    raise FileNotFoundError("Không tìm thấy file pikachu.jpg")
```

```python

# 1. Thay đổi độ sáng
brightness = 50
bright_img = cv2.add(img, brightness)

dim = -200
dim_img = cv2.add(img, dim)
```

```python

# 2. Thay đổi độ tương phản
contrast = 1.5
contrast_img = cv2.convertScaleAbs(img, alpha=contrast, beta=0)
```

```python

# 3. Biến đổi âm bản
negative_img = 255 - img
```

```python

# 4. Cắt ngưỡng
threshold_gray = cv2.threshold(
    cv2.cvtColor(img, cv2.COLOR_BGR2GRAY),
    128, 255, cv2.THRESH_BINARY
)[1]
threshold_img = cv2.cvtColor(threshold_gray, cv2.COLOR_GRAY2BGR)
```

```python

def resize(img, size=(300, 300)):
    return cv2.resize(img, size)

def show(img, title):
    plt.figure(figsize=(6,3))
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.axis("off")
    plt.show()
```

```python

# Hiển thị câu 1
img_cau1 = np.hstack((
    resize(img),
    resize(bright_img),
    resize(dim_img)
))
show(img_cau1, "1. Origin | Brightness | Dim")
```

```python

# Hiển thị câu 2
img_cau2 = np.hstack((
    resize(img),
    resize(contrast_img)
))
show(img_cau2, "2. Origin | Contrast")
```

```python

# Hiển thị câu 3
img_cau3 = np.hstack((
    resize(img),
    resize(negative_img)
))
show(img_cau3, "3. Origin | Negative")
```

```python

# Hiển thị câu 4
img_cau4 = np.hstack((
    resize(img),
    resize(threshold_img)
))
show(img_cau4, "4. Origin | Threshold")
```
