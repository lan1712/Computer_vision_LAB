```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Cu hnh  hin th nh to v r hn
plt.rcParams['figure.figsize'] = [10, 5]

# c nh
img = cv2.imread("pikachu.jpg")

# Lu : OpenCV c nh h BGR, Matplotlib hin th h RGB. 
# Tuy nhin bn ang lm vic trn nh xm (Gray), nn chng ta chuyn i sang Gray.
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Hm h tr hin th nh nhanh bng Matplotlib
def show_image(image, title="Image", cmap_type='gray'):
    plt.imshow(image, cmap=cmap_type)
    plt.title(title)
    plt.axis('off') # n trc ta 
    plt.show()

# Hin th nh gc (Gray)
show_image(gray, "Original Grayscale")
```

```python
# Lc trung bnh: Lm m bng vng ln cn (5x5)
blur = cv2.blur(gray, (5, 5))
show_image(blur, "1. Average Blur")
```

```python
# Lc Gaussian: Lm m t nhin hn
gaussian_blur = cv2.GaussianBlur(gray, (5, 5), 0)
show_image(gaussian_blur, "2. Gaussian Blur")
```

```python
# nh ngha kernel lm sc nt
kernel_sharpen = np.array([[0, -1, 0],
                           [-1, 5,-1],
                           [0, -1, 0]])

# p dng filter2D
sharpen = cv2.filter2D(gray, -1, kernel_sharpen)
show_image(sharpen, "3. Sharpened")
```
